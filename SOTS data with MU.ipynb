```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
mpl.rc('font', family='Segoe UI Symbol')
```

```python
sots_df = pd.read_csv('C:/Users/she384/Documents/SOTS Nutrient Variability Paper/Data/Collated_Data.csv')
```

```python
sots_df.head()
```

```python
sots_2015_df = sots_df.loc[(sots_df['Voyage'] == 'in2015_v01') & (sots_df['Deployment'] == 5)].dropna(subset=['NOx (uM)'])
sots_2016_df = sots_df.loc[(sots_df['Voyage'] == 'in2016_v02') & (sots_df['Deployment'] == 2)].dropna(subset=['NOx (uM)'])
sots_2017_df = sots_df.loc[(sots_df['Voyage'] == 'in2017_v02') & (sots_df['Deployment'] == 5)].dropna(subset=['NOx (uM)'])
sots_2018_df = sots_df.loc[(sots_df['Voyage'] == 'in2018_v02') & (sots_df['Deployment'] == 4)].dropna(subset=['NOx (uM)'])
sots_2019_df = sots_df.loc[(sots_df['Voyage'] == 'in2019_v02') & (sots_df['Deployment'] == 2)].dropna(subset=['NOx (uM)'])
```

```python
      
lower_limit_2015 = [(x-(x*0.015)) for x in list(sots_2015_df['NOx (uM)'])]
upper_limit_2015 = [(x+(x*0.015)) for x in list(sots_2015_df['NOx (uM)'])]

lower_limit_2016 = [(x-(x*0.015)) for x in list(sots_2016_df['NOx (uM)'])]
upper_limit_2016 = [(x+(x*0.015)) for x in list(sots_2016_df['NOx (uM)'])]

lower_limit_2017 = [(x-(x*0.015)) for x in list(sots_2017_df['NOx (uM)'])]
upper_limit_2017 = [(x+(x*0.015)) for x in list(sots_2017_df['NOx (uM)'])]

lower_limit_2018 = [(x-(x*0.015)) for x in list(sots_2018_df['NOx (uM)'])]
upper_limit_2018 = [(x+(x*0.015)) for x in list(sots_2018_df['NOx (uM)'])]


lower_limit_2019 = [(x-(x*0.015)) for x in list(sots_2019_df['NOx (uM)'])]
upper_limit_2019 = [(x+(x*0.015)) for x in list(sots_2019_df['NOx (uM)'])]

plt.figure(figsize=(12, 15))

plt.plot(sots_2015_df['NOx (uM)'], sots_2015_df['Pressure'], lw=0.5, alpha=0.8, marker='o', ms=1, label='2015', color='#5d87e3')
plt.plot(sots_2016_df['NOx (uM)'], sots_2016_df['Pressure'], lw=0.5, alpha=0.8, marker='o', ms=1, label='2016', color='#ed9a2f')
plt.plot(sots_2017_df['NOx (uM)'], sots_2017_df['Pressure'], lw=0.5, alpha=0.8, marker='o', ms=1, label='2017', color='#0ec45f')
plt.plot(sots_2018_df['NOx (uM)'], sots_2018_df['Pressure'], lw=0.5, alpha=0.8, marker='o', ms=1, label='2018', color='#8e22e0')
plt.plot(sots_2019_df['NOx (uM)'], sots_2019_df['Pressure'], lw=0.5, alpha=0.8, marker='o', ms=1, label='2019', color='#e02255')

plt.fill_betweenx(sots_2015_df['Pressure'], lower_limit_2015, upper_limit_2015, color='#5d87e3', alpha=0.1, label='2015 95% CI')
plt.fill_betweenx(sots_2016_df['Pressure'], lower_limit_2016, upper_limit_2016, color='#ed9a2f', alpha=0.2, label='2016 95% CI')
plt.fill_betweenx(sots_2017_df['Pressure'], lower_limit_2017, upper_limit_2017, color='#0ec45f', alpha=0.1, label='2017 95% CI')
plt.fill_betweenx(sots_2018_df['Pressure'], lower_limit_2018, upper_limit_2018, color='#8e22e0', alpha=0.1, label='2018 95% CI')
plt.fill_betweenx(sots_2019_df['Pressure'], lower_limit_2019, upper_limit_2019, color='#e02255', alpha=0.1, label='2019 95% CI')


plt.gca().invert_yaxis()
plt.grid(alpha=0.2, zorder=-1)
plt.title('NOx Concentration at SOTS site: 2015-2019', fontsize=22)
plt.xlabel('NOx Concentration (uM)', fontsize=17)
plt.ylabel('Pressure (db)', fontsize=17)
plt.legend(fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(5000, 0)
```
